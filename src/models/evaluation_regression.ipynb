```python
plt.figure(figsize=(10, 6))
plt.plot(scratch_linear_regression_metrics["Loss History"], label="Linear")
plt.plot(scratch_poly_regression_metrics["Loss History"], label="Polynomial")
plt.plot(scratch_multiple_regression_metrics["Loss History"], label="Multiple")

plt.title("Training Loss Comparison (MSE)")
plt.xlabel("Epoch / Iteration")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.plot(mlp_regression_metrics["Loss History"], label="MLP Train Loss")
plt.plot(mlp_regression_metrics["Val History"], label="MLP Validation Loss", linestyle="--")

plt.title("MLP Training Loss (Scaled MSE)")
plt.xlabel("Epoch")
plt.ylabel("Scaled MSE")
plt.legend()
plt.grid(True)

plt.show()
```

```python
models = [
    "Linear (Scratch)",
    "Linear (Sklearn)",
    "Polynomial (Scratch)",
    "Polynomial (Sklearn)",
    "Multiple (Scratch)",
    "Multiple (Sklearn)",
    "MLP (Scratch)"
]

train_mse = [
    scratch_linear_regression_metrics["Train MSE"],
    sklearn_linear_regression_metrics["Train MSE"],
    scratch_poly_regression_metrics["Train MSE"],
    sklearn_poly_regression_metrics["Train MSE"],
    scratch_multiple_regression_metrics["Train MSE"],
    sklearn_multiple_regression_metrics["Train MSE"],
    mlp_regression_metrics["Train MSE"],
]

test_mse = [
    scratch_linear_regression_metrics["Test MSE"],
    sklearn_linear_regression_metrics["Test MSE"],
    scratch_poly_regression_metrics["Test MSE"],
    sklearn_poly_regression_metrics["Test MSE"],
    scratch_multiple_regression_metrics["Test MSE"],
    sklearn_multiple_regression_metrics["Test MSE"],
    mlp_regression_metrics["Test MSE"],
]

x = np.arange(len(models))
width = 0.35

plt.figure(figsize=(14, 6))

plt.bar(x - width/2, train_mse, width, label="Train MSE")
plt.bar(x + width/2, test_mse, width, label="Test MSE")

plt.xticks(x, models, rotation=30)
plt.ylabel("MSE")
plt.title("Model Comparison (Train vs Test MSE)")
plt.legend()
plt.grid(axis="y")

plt.tight_layout()
plt.show()
```
