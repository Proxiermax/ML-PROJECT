```python
# --- Accuracy Comparison ---
models = [
    "Logistic Reg.",
    "Perceptron",
    "MLP",
    "Decision Tree",
    "Random Forest",
    "SVM",
    "KNN (Custom)",
]

accuracy_values = [
    logistic_metrics["accuracy"],
    perceptron_metrics["accuracy"],
    mlp_metrics["accuracy"],
    decision_tree_metrics["accuracy"],
    random_forest_metrics["accuracy"],
    svm_metrics["accuracy"],
    custom_metrics["accuracy"],
]

plt.figure(figsize=(12, 6))
bars = plt.bar(models, accuracy_values, color="steelblue")
plt.ylabel("Accuracy")
plt.title("Classification Model Comparison — Accuracy")
plt.ylim(0, 1.05)
plt.xticks(rotation=30, ha="right")
for bar, val in zip(bars, accuracy_values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01,
             f"{val:.4f}", ha="center", va="bottom", fontsize=9)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
# --- Multi-Metric Comparison (Precision, Recall, F1) ---
metrics_keys = ["precision", "recall", "f1"]
x = np.arange(len(models))
width = 0.25

all_metrics = [logistic_metrics, perceptron_metrics, mlp_metrics,
               decision_tree_metrics, random_forest_metrics,
               svm_metrics, custom_metrics]

plt.figure(figsize=(14, 6))
for i, key in enumerate(metrics_keys):
    values = [m[key] for m in all_metrics]
    plt.bar(x + i * width, values, width, label=key.capitalize())

plt.xticks(x + width, models, rotation=30, ha="right")
plt.ylabel("Score")
plt.title("Classification Model Comparison — Precision / Recall / F1")
plt.ylim(0, 1.05)
plt.legend()
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```
