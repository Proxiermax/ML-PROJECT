```python
mse_labels = [
    "Scratch Train",
    "Scratch Test",
    "Sklearn Train",
    "Sklearn Test"
]

mse_values = [
    lr_metrics["Scratch Train MSE"],
    lr_metrics["Scratch Test MSE"],
    lr_metrics["Sklearn Train MSE"],
    lr_metrics["Sklearn Test MSE"]
]

plt.figure()
plt.bar(mse_labels, mse_values)
plt.title("MSE Comparison")
plt.xticks(rotation=45)
plt.ylabel("MSE")
plt.show()
```
