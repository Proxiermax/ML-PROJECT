```python
import matplotlib.pyplot as plt
import importlib
import numpy as np
import pickle
from pathlib import Path
```

```python
PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / "models").exists():
    PROJECT_ROOT = Path(__file__).resolve().parents[2] if '__file__' in dir() else Path.cwd()
    while not (PROJECT_ROOT / "models").exists() and PROJECT_ROOT != PROJECT_ROOT.parent:
        PROJECT_ROOT = PROJECT_ROOT.parent

SCRATCH_MODEL_DIR = PROJECT_ROOT / "models" / "classification" / "scratch"

def load_model(model_name):
    path = SCRATCH_MODEL_DIR / model_name / "model.pkl"
    if path.exists():
        with open(path, "rb") as f:
            pkg = pickle.load(f)
        print(f"Loaded saved model from {path}")
        return pkg
    else:
        print(f"No saved model found at {path} — run training cell")
        return None
```

```python
models = [
    "Logistic Reg.",
    "Perceptron",
    "MLP",
    "Decision Tree",
    "Random Forest",
    "SVM",
    "KNN (Custom)",
    "XGBoost",
]

accuracy_values = [
    logistic_metrics["accuracy"],
    perceptron_metrics["accuracy"],
    mlp_metrics["accuracy"],
    decision_tree_metrics["accuracy"],
    random_forest_metrics["accuracy"],
    svm_metrics["accuracy"],
    custom_metrics["accuracy"],
    xgboost_metrics["accuracy"],
]

plt.figure(figsize=(12, 6))
bars = plt.bar(models, accuracy_values, color="steelblue")
plt.ylabel("Accuracy")
plt.title("Classification Model Comparison — Accuracy")
plt.ylim(0, 1.05)
plt.xticks(rotation=30, ha="right")
for bar, val in zip(bars, accuracy_values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01,
             f"{val:.4f}", ha="center", va="bottom", fontsize=9)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
metrics_keys = ["precision", "recall", "f1"]
x = np.arange(len(models))
width = 0.25

all_metrics = [logistic_metrics, perceptron_metrics, mlp_metrics,
               decision_tree_metrics, random_forest_metrics,
               svm_metrics, custom_metrics, xgboost_metrics]

plt.figure(figsize=(14, 6))
for i, key in enumerate(metrics_keys):
    values = [m[key] for m in all_metrics]
    plt.bar(x + i * width, values, width, label=key.capitalize())

plt.xticks(x + width, models, rotation=30, ha="right")
plt.ylabel("Score")
plt.title("Classification Model Comparison — Precision / Recall / F1")
plt.ylim(0, 1.05)
plt.legend()
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
# --- Specificity / True Negative Rate Comparison ---
specificity_values = [m["specificity"] for m in all_metrics]

plt.figure(figsize=(12, 6))
bars = plt.bar(models, specificity_values, color="darkorange")
plt.ylabel("Specificity (TNR)")
plt.title("Classification Model Comparison — Specificity / True Negative Rate")
plt.ylim(0, 1.05)
plt.xticks(rotation=30, ha="right")
for bar, val in zip(bars, specificity_values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01,
             f"{val:.4f}", ha="center", va="bottom", fontsize=9)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
# --- ROC Curve & AUC ---
from sklearn.metrics import roc_curve, auc

roc_models = [
    "Logistic Reg.",
    "Perceptron",
    "MLP",
    "Decision Tree",
    "Random Forest",
    "SVM",
    "KNN (Custom)",
    "XGBoost",
]
roc_metrics = [logistic_metrics, perceptron_metrics, mlp_metrics,
               decision_tree_metrics, random_forest_metrics,
               svm_metrics, custom_metrics, xgboost_metrics]

plt.figure(figsize=(10, 8))
auc_scores = {}

for name, m in zip(roc_models, roc_metrics):
    y_true = m["y_test"]
    y_scores = m["y_scores"]
    fpr, tpr, _ = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)
    auc_scores[name] = roc_auc
    plt.plot(fpr, tpr, linewidth=2, label=f"{name} (AUC={roc_auc:.4f})")

plt.plot([0, 1], [0, 1], "k--", linewidth=1, label="Random (AUC=0.5)")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve — All Classification Models")
plt.legend(loc="lower right", fontsize=9)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- AUC Bar Chart ---
auc_names = list(auc_scores.keys())
auc_vals = list(auc_scores.values())

plt.figure(figsize=(12, 6))
bars = plt.bar(auc_names, auc_vals, color="mediumseagreen")
plt.ylabel("AUC Score")
plt.title("AUC Score Comparison — All Classification Models")
plt.ylim(0, 1.05)
plt.xticks(rotation=30, ha="right")
for bar, val in zip(bars, auc_vals):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01,
             f"{val:.4f}", ha="center", va="bottom", fontsize=9)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
# --- Performance Curves: Training Loss & Accuracy ---
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# --- Left: Loss curves ---
ax = axes[0]
ax.plot(logistic_model.loss_history, label="Logistic Reg.")
ax.plot(perceptron_model.loss_history, label="Perceptron")
ax.plot(mlp_model.loss_history, label="MLP")
ax.plot(svm_model.loss_history, label="SVM")
ax.plot(xgboost_model.loss_history, label="XGBoost")
ax.set_xlabel("Epoch / Iteration")
ax.set_ylabel("Loss")
ax.set_title("Training Loss Curves")
ax.legend()
ax.grid(True)

# --- Right: Accuracy curves ---
ax2 = axes[1]
ax2.plot(perceptron_model.accuracy_history, label="Perceptron")
ax2.plot(mlp_model.accuracy_history, label="MLP")
ax2.set_xlabel("Epoch / Iteration")
ax2.set_ylabel("Accuracy")
ax2.set_title("Training Accuracy Curves")
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
