```python
import matplotlib.pyplot as plt
import importlib
import numpy as np
import pickle
from pathlib import Path
```

```python
PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / "models").exists():
    PROJECT_ROOT = Path(__file__).resolve().parents[2] if '__file__' in dir() else Path.cwd()
    while not (PROJECT_ROOT / "models").exists() and PROJECT_ROOT != PROJECT_ROOT.parent:
        PROJECT_ROOT = PROJECT_ROOT.parent

SCRATCH_MODEL_DIR = PROJECT_ROOT / "models" / "classification" / "scratch"
LIB_MODEL_DIR = PROJECT_ROOT / "models" / "classification" / "lib"

def load_classification_model(model_name, source="scratch"):
    base = SCRATCH_MODEL_DIR if source == "scratch" else LIB_MODEL_DIR
    path = base / model_name / "model.pkl"
    if path.exists():
        with open(path, "rb") as f:
            pkg = pickle.load(f)
        print(f"Loaded {source}/{model_name} from {path}")
        return pkg
    else:
        print(f"No saved model at {path} — run training cell")
        return None
```

```python
models = [
    "Logistic Reg. (Scratch)", "Logistic Reg. (Sklearn)",
    "Perceptron (Scratch)", "Perceptron (Sklearn)",
    "MLP (Scratch)", "MLP (Sklearn)",
    "Decision Tree (Scratch)", "Decision Tree (Sklearn)",
    "Random Forest (Scratch)", "Random Forest (Sklearn)",
    "SVM (Scratch)", "SVM (Sklearn)",
    "KNN (Scratch)", "KNN (Sklearn)",
    "XGBoost (Scratch)",
]

accuracy_values = [
    scratch_logistic_metrics["accuracy"], sklearn_logistic_metrics["accuracy"],
    scratch_perceptron_metrics["accuracy"], sklearn_perceptron_metrics["accuracy"],
    scratch_mlp_metrics["accuracy"], sklearn_mlp_metrics["accuracy"],
    scratch_decision_tree_metrics["accuracy"], sklearn_decision_tree_metrics["accuracy"],
    scratch_random_forest_metrics["accuracy"], sklearn_random_forest_metrics["accuracy"],
    scratch_svm_metrics["accuracy"], sklearn_svm_metrics["accuracy"],
    scratch_custom_metrics["accuracy"], sklearn_custom_metrics["accuracy"],
    scratch_xgboost_metrics["accuracy"],
]

colors = ["steelblue" if "(Scratch)" in m else "darkorange" for m in models]

plt.figure(figsize=(16, 6))
bars = plt.bar(models, accuracy_values, color=colors)
plt.ylabel("Accuracy")
plt.title("Classification Model Comparison — Accuracy (Scratch vs Sklearn)")
plt.ylim(0, 1.05)
plt.xticks(rotation=45, ha="right")
for bar, val in zip(bars, accuracy_values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01,
             f"{val:.4f}", ha="center", va="bottom", fontsize=8)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
metrics_keys = ["precision", "recall", "f1"]
x = np.arange(len(models))
width = 0.25

all_metrics = [
    scratch_logistic_metrics, sklearn_logistic_metrics,
    scratch_perceptron_metrics, sklearn_perceptron_metrics,
    scratch_mlp_metrics, sklearn_mlp_metrics,
    scratch_decision_tree_metrics, sklearn_decision_tree_metrics,
    scratch_random_forest_metrics, sklearn_random_forest_metrics,
    scratch_svm_metrics, sklearn_svm_metrics,
    scratch_custom_metrics, sklearn_custom_metrics,
    scratch_xgboost_metrics,
]

plt.figure(figsize=(18, 6))
for i, key in enumerate(metrics_keys):
    values = [m[key] for m in all_metrics]
    plt.bar(x + i * width, values, width, label=key.capitalize())

plt.xticks(x + width, models, rotation=45, ha="right")
plt.ylabel("Score")
plt.title("Classification Model Comparison — Precision / Recall / F1 (Scratch vs Sklearn)")
plt.ylim(0, 1.05)
plt.legend()
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
# --- Confusion Matrix Summary Table ---
print(f"{'Model':<30} {'TP':>6} {'FP':>6} {'FN':>6} {'TN':>6} {'Accuracy':>10} {'F1':>8}")
print("-" * 72)
for name, m in zip(models, all_metrics):
    print(f"{name:<30} {m['tp']:>6} {m['fp']:>6} {m['fn']:>6} {m['tn']:>6} {m['accuracy']:>10.4f} {m['f1']:>8.4f}")
```

```python
# --- Specificity / True Negative Rate Comparison ---
specificity_values = [m["specificity"] for m in all_metrics]

colors = ["steelblue" if "(Scratch)" in m else "darkorange" for m in models]

plt.figure(figsize=(16, 6))
bars = plt.bar(models, specificity_values, color=colors)
plt.ylabel("Specificity (TNR)")
plt.title("Classification Model Comparison — Specificity (Scratch vs Sklearn)")
plt.ylim(0, 1.05)
plt.xticks(rotation=45, ha="right")
for bar, val in zip(bars, specificity_values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01,
             f"{val:.4f}", ha="center", va="bottom", fontsize=8)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
# --- ROC Curve & AUC ---
from sklearn.metrics import roc_curve, auc

roc_models = models  # reuse the full scratch+sklearn model list
roc_metrics = all_metrics

plt.figure(figsize=(10, 8))
auc_scores = {}

for name, m in zip(roc_models, roc_metrics):
    y_true = m["y_test"]
    y_scores = m["y_scores"]
    fpr, tpr, _ = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)
    auc_scores[name] = roc_auc
    linestyle = "-" if "(Scratch)" in name else "--"
    plt.plot(fpr, tpr, linewidth=2, linestyle=linestyle, label=f"{name} (AUC={roc_auc:.4f})")

plt.plot([0, 1], [0, 1], "k--", linewidth=1, label="Random (AUC=0.5)")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve — All Classification Models (Scratch vs Sklearn)")
plt.legend(loc="lower right", fontsize=8)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- AUC Bar Chart ---
auc_names = list(auc_scores.keys())
auc_vals = list(auc_scores.values())
colors = ["steelblue" if "(Scratch)" in n else "darkorange" for n in auc_names]

plt.figure(figsize=(16, 6))
bars = plt.bar(auc_names, auc_vals, color=colors)
plt.ylabel("AUC Score")
plt.title("AUC Score Comparison — All Classification Models (Scratch vs Sklearn)")
plt.ylim(0, 1.05)
plt.xticks(rotation=45, ha="right")
for bar, val in zip(bars, auc_vals):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01,
             f"{val:.4f}", ha="center", va="bottom", fontsize=8)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```
